### Arranging & Renaming the dataset

```python
data=data[["iyear","imonth","iday","country_txt","region_txt","provstate","city",
       "latitude","longitude","location","summary","attacktype1_txt","targtype1_txt",
       "gname","motive","weaptype1_txt","nkill","nwound","addnotes"]]
```

```python

data.rename(columns={"iyear":"Year","imonth":"Month","iday":"Day","country_txt":"Country",
                   "region_txt":"Region","provstate":"Province/State","city":"City",
                   "latitude":"Latitude","longitude":"Longitude","location":"Location",
                   "summary":"Summary","attacktype1_txt":"Attack_Type","targtype1_txt":"Target_Type",
                   "gname":"Group","motive":"Motive","weaptype1_txt":"Weapon_Type",
                   "nkill":"Killed","nwound":"Wounded","addnotes":"Add_Notes"}, inplace=True)
```

```python
data.head()
```

### Checking & Removing null values

```python
data.isnull().sum()
```

```python
data.shape
```

```python
data_new = data.dropna(axis=0)
```

```python
data_new.head()
```

```python
data_new.shape
```

```python
data_new.info()
```

### Number of Terrorist Activities each Year

```python
year = data_new['Year'].unique()
years_count = data_new['Year'].value_counts(dropna = False).sort_index()
plt.figure(figsize = (15,10))
sns.barplot(x = year,
            y = years_count,
            palette = 'muted')
plt.xticks(rotation = 50)
plt.xlabel('Years', fontsize = 20)
plt.ylabel('Number of attacks each year', fontsize = 20)
plt.title('Attacks In Years', fontsize = 25)
```

### Terrorist Activities by Region in each Year through Area Plot

```python
pd.crosstab(data_new.Year, data_new.Region).plot(kind='area',stacked=False,figsize=(20,10))
plt.title('Terrorist Activities by Region in each Year', size = 25)
plt.xlabel('Year' , size = 20)
plt.ylabel('Number of Attacks', size = 20)
plt.show()
```

### Most Affected Cities

```python
data_new['City'].value_counts().to_frame().sort_values('City',axis=0,ascending=False).head(10).plot(kind='bar',figsize=(20,10), color='indianred')
plt.xticks(rotation = 50)
plt.xlabel("City",fontsize = 20)
plt.ylabel("Number of attack",fontsize = 20)
plt.title("Most Affected Cities",fontsize = 25)
plt.show()
```

### Top Terrorist Groups

```python
data_new['Group'].value_counts()[:10]
```

```python
data_new[['Group','Killed']].groupby(['Group'],axis=0).sum().drop('Unknown').sort_values('Killed',ascending=False).head(10).plot(kind='bar', color='indianred',figsize=(20,10))
plt.title("Top Terrorist Groups With Respect To People Killed",fontsize = 25)
plt.xlabel("Group Name",fontsize = 20)
plt.ylabel("Number Of People Killed",fontsize = 20)
plt.show()
```

### Most Affected Countries

```python
plt.subplots(figsize=(20,10))
plt.bar(data_new['Country'].value_counts()[:10].index,data_new['Country'].value_counts()[:10].values)
plt.title('Top Countries Affected', fontsize = 25)
plt.xlabel('Countries', fontsize = 20)
plt.ylabel('Count', fontsize = 20)
plt.xticks(rotation = 50)
plt.show()
```

### Country wise Casualities

```python
lis=[]
total=[]
for i in data_new['Country'].value_counts().index:
    Total = data_new.loc[data_new['Country'] == i, 'Killed'].sum()
    lis.append([i,Total])
lis=sorted(lis,key=lambda x:x[1],reverse=True)
print(*lis[:20])   #country wise casuality over the years from 1970 to 2017
```

```python
country=[]
casuality=[]
for i in lis[:11]:
    country.append(i[0])
    casuality.append(i[1])
    
fig=plt.figure(figsize=(20,10))
plt.title("Country wise Casualities (1970-2017)",fontsize=25)
plt.xlabel("Country",size=25)
plt.ylabel("Casuality",size=25)
sns.barplot(x=country,y=casuality)
plt.show()
```

### Conclusion :

```python
print("Country with the most attacks:",data_new['Country'].value_counts().idxmax())
print("City with the most attacks:",data_new['City'].value_counts().index[0])
print("Region with the most attacks:",data_new['Region'].value_counts().idxmax())
print("Year with the most attacks:",data_new['Year'].value_counts().idxmax())
print("Month with the most attacks:",data_new['Month'].value_counts().idxmax())
print("Group with the most attacks:",data_new['Group'].value_counts().index[1])
```
